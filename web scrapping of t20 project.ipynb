# WEB SCRAPPING

BeautifulSoap ->  Purpose: Beautiful Soup is used for parsing and navigating HTML or XML documents, making it a powerful tool for web scraping and extracting specific data from web pages.

Features: It provides methods to search, filter, and extract data from HTML or XML documents. 

Requests -> Purpose: The requests library is used for making HTTP requests to retrieve web pages, data, and interact with web services.

Features: It allows you to send GET, POST, PUT, DELETE, and other types of HTTP requests. It handles the communication between your Python script and web servers, providing the raw HTML content of web pages or data from web APIs.

```python
from bs4 import BeautifulSoup
import requests as r
```

#Extract match result data from ESPN

```python
url = 'https://www.espncricinfo.com/records/tournament/team-match-results/icc-men-s-t20-world-cup-2022-23-14450'
page_res = r.get(url)      #making request to web server and handling the respones
```

```python
page_res     #response 200 which the serer accepts the request
```

```python
soup = BeautifulSoup(page_res.text, 'html')
```

```python
col_title = soup.find('tr')     #Extracting column names
col_title
```

```python
match_res_title = [title.text.strip() for title in col_title]   #iterating into title column in table using list comprehension
```

```python
match_res_title
```

Naming dataframe with names in variable match_res_title

```python
import pandas as pd
df = pd.DataFrame(columns = match_res_title)
df
```

```python
l=0
temp = soup.find_all('tr')
for row,row_data in enumerate(temp[1:]):  #start the loop from 1 to eliminate the header column
    match_res_data = [data.text.strip() for data in row_data]
    
    df.loc[l] = match_res_data
    l+=1
    
```

```python
df
```

```python
df.to_csv('matchs_and_results.csv')
```
